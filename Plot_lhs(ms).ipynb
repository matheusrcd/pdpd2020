```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import e

plt.rcParams['figure.figsize'] = [14, 5]


x_hss = np.linspace(0, 62.5, num=100)
x_ppss = np.linspace(62, 500)
x_rh = np.linspace(62, 500)
x_rl = np.linspace(0, 63)

plt.plot(x_hss, 0.0238492+1.2092239340844e-23* e**(0.795353*x_hss), linewidth=8)
plt.plot(x_ppss, x_ppss*0 + 12, linewidth=8)
plt.plot(x_rh, (0.0020986* x_rh**(1.94613))-1.91028, linewidth=2, color='r')

plt.plot(x_rl, 0.185187+6.1625900511865e-30* e**(1.06897*x_rl), linewidth=2, color='r')


plt.gca().set_facecolor("#3b3b3b")
plt.yscale('log')
plt.axhspan(0, 12, facecolor='#7bc46a')
plt.axvspan(0, 62, facecolor='#3b3b3b')
plt.axhspan(0, 0.029, facecolor='#7bc46a')
plt.xlabel('Massa (Gev)')
plt.margins(y=1, x =0)
plt.ylabel('LHS')

plt.title("Acoplamento do higgs em função da massa de matéria escura")

plt.show()
```
